```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=15)
```

```python
class A:
  def perceptron_G(self,x,y):
    w1=w2=b=1
    lr=0.001
    for j in range(1000):
      for i in range(x.shape[0]):
        z=w1*x[i][0]+w2*x[i][1]+b
        if z*y[i]<0:
          w1=w1+lr*x[i][0]*y[i]
          w2=w2+lr*x[i][1]*y[i]
          b=b+lr*y[i]
    return w1,w2,b
```

```python
m=A()
```

```python
w1,w2,b=(m.perceptron_G(x,y))
```

```python
print(w1,w2,b,end='')
```

```python
#now make a line for ploting
#code below
#let's slope 'm' & intercept 'c'
#let's define ...coefficeint of 'x',& coefficient of 'y' is -> coef_'x' ='w1',coef_'y'='w2'
#slope(m)=-(coef_'x')/(coef_'y')
m=-(w1/w2)
#intercept_'c'='b'
# 'c'=-(c/coe_'y)
c=-(b/w2)
```

```python
print(m,c,end='')
```

```python
x_input=np.linspace(-4,3,100)
y_input=(m*x_input+c)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x_input,y_input)
plt.scatter(x[:,0],x[:,1],c=y,cmap="RdBu",s=100)
```

```python
x1=[]
x2=[]
for i in range(len(x)):
  for j in range(len(x)):
    x1.append(x[i][0])
    x2.append(x[i][1])
```

```python
def predict(a,a1):
  p=w1*a+w2*a1+b
  if p>0:
    return 1
  else:
    return 0
```
